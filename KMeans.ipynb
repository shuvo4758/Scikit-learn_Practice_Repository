# KMeans Clustering

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# sample data
data = {
    'Customer': ['Shuvo', 'Provat', 'Akash', 'Robin', 'Saidur', 'Aquib'],
    'Age': [20, 30, 40, 22, 38, 25],
    'Spending': [100, 200, 300, 110, 290, 130]
}

df = pd.DataFrame(data)

# selecting usefull columns
X = df[['Age', 'Spending']]

# creating model
model = KMeans(n_clusters=2, random_state=42, n_init=10)

# fit & predict
df['Group'] = model.fit_predict(X)

# plotting
plt.figure(figsize=(6, 5))
for group in df['Group'].unique():    # either 0 or 1
  group_data = df[df['Group'] == group]   # masking
  plt.scatter(group_data['Age'], group_data['Spending'], label=f'Group {group}')

plt.xlabel('Age')
plt.ylabel('Spending')
plt.title('Customer Groups (KMeans)')
plt.legend()
plt.grid(True)
plt.show()

print(df)
```

```
from sklearn.linear_model import LinearRegression

X = [[1], [2], [3], [4], [5]]   # hour studied
y = [40, 50, 65, 75, 90]        # marks

model = LinearRegression()

model.fit(X, y)

hours = float(input("Enter how many hours you studied :"))

predicted_marks = model.predict([[hours]])

print(f"Based on your {hours} hours of study, you may score around {predicted_marks} marks")
```

```
from sklearn.linear_model import LogisticRegression

X = [[1], [2], [3], [4], [5]]   # hour studied
y = [0, 0, 1, 1, 1]        # marks

model = LogisticRegression()

model.fit(X, y)

hours = float(input("Enter how many hours you studied :"))

result = model.predict([[hours]])

if result == 1:
  print("You may pass the exam")
else:
  print("You may fail the exam")
```

```
from sklearn.neighbors import KNeighborsClassifier

X = [           # [ weight in grams, size in cm ]
    [180, 7],
    [200, 7.5],
    [250, 8],
    [300, 8.5],
    [350, 9],
    [400, 9.5]
    ]
y = [0, 0, 0, 1, 1, 1]        # 0=Apple, 1=Orange

model = KNeighborsClassifier(n_neighbors=3)

model.fit(X, y)

weight = float(input("Enter weight in grams :"))
size = float(input("Enter size in cm :"))

prediction = model.predict([[weight, size]])

if prediction == 0:
  print("It is likely an Apple")
else:
  print("It is likely an Orange")
```

```
from sklearn.tree import DecisionTreeClassifier

X = [           # [size in cm, color shade]
    [7, 2],
    [8, 3],
    [9, 8],
    [10, 9]
    ]
y = [0, 0, 1, 1]        # 0=Apple, 1=Orange

model = DecisionTreeClassifier()

model.fit(X, y)

size = float(input("Enter size in cm :"))
shade = float(input("Enter color shade between(0-10) :"))

prediction = model.predict([[size, shade]])[0]    # passing 0 to acheive the actual value, not the list

if prediction == 0:
  print("It is likely an Apple")
else:
  print("It is likely an Orange")
```
